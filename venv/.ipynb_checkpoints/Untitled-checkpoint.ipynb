```python

#Captures Images via Webcam and Performs Face Recognition
face_recognizer = cv2.face.LBPHFaceRecognizer_create()
face_recognizer.read('trainingData.yml')#Load saved training data

name={0:"Rahul",1:"Daya",2:"Anees",3:"Sangu",4:"Avinash",5:"Vishal",6:"Maaz",7:"Sridhar",8:"Shubham",9:"Sandy",10:"Srinu",11:"Joshi",12:"Mayur",13:"Diksha",14:"Reshma",15:"Guggu",16:"Mahesh",17:"Junior",18:"Senior",19:"Manoj",20:"Anand",21:"Ajji",22:"Thanu",23:"",24:"pallavi",25:"parvathi",26:"Anushka",27:"UNKNOWN"}


cap=cv2.VideoCapture(0)

while True:
    ret,test_img=cap.read()# captures frame and returns boolean value and captured image
    faces_detected,gray_img=fr.faceDetection(test_img)



    for (x,y,w,h) in faces_detected:
      cv2.rectangle(test_img,(x,y),(x+w,y+h),(255,0,0),thickness=7)

    resized_img = cv2.resize(test_img, (1000, 700))
    cv2.imshow('face detection Tutorial ',resized_img)
    cv2.waitKey(10)


    for face in faces_detected:
        (x,y,w,h)=face
        roi_gray=gray_img[y:y+w, x:x+h]
        label,confidence=face_recognizer.predict(roi_gray)#predicting the label of given image
        print("confidence:",confidence)
        print("label:",label)
        fr.draw_rect(test_img,face)
        predicted_name=name[label]
        if confidence < 39:
           fr.put_text(test_img,predicted_name,x,y)


    resized_img = cv2.resize(test_img, (1000, 700))
    cv2.imshow('Face Recognition System ',resized_img)
    if cv2.waitKey(10)== ord("q"):#wait until 'q' key is pressed
        break


cap.release()
cv2.destroyAllWindows

```
